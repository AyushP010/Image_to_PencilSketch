```python
import cv2
import numpy as np
def pencil_sketch(image_path, output_path):
  cv2.imread(image_path)
```

```python
image = cv2.imread('dog.jpg')
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
inv_gray = 255 - gray
```

```python
blur = cv2.GaussianBlur(inv_gray, (21, 21), 0)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
# Sketch by dividing the grayscale image by the inverted blurred image
#sketch_image = cv2.divide(gray_image, inverted_image, scale=111.0)
```

```python
pencil_sketch = cv2.divide(gray, 255 - blur, scale=256)
```

```python
cv2_imshow(pencil_sketch)
```

```python
#save pencil sketch
cv2.imwrite('pencil_sketch.jpg', pencil_sketch)
```
